```python
x=pd.DataFrame()
x["Delta T Compression"]=dtco
x["Delts T Shear"]=dts
x["Depth"]=depth
x["Gamma"]=ecgr
x["neutronPor"]=neuPor
```

```python
x=x[(x["Depth"]>=2643.08) & (x["Depth"]<=2780.80)]
```

```python
dens=pd.DataFrame()
```

```python
dens["Depth"]=depth2
dens["Density"]=density
dens["mudcake"]=mudcakeden
dens["denPor"]=denPor
```

```python
dens=dens[(dens["Depth"]>=2643.08) & (dens["Depth"]<=2780.80)]
```

```python
merge=pd.merge(x,dens,on=["Depth"])
```

```python
merge.rename(columns={'mudcake': 'Mudcake Density(g/cm3)', 'Density': 'Density(g/cm3)','Depth':'Depth(m)'}, inplace=True)
```

```python
por=list()
for i in range(len(merge['denPor'])):
    if merge['denPor'][i]<merge['neutronPor'][i] and merge['denPor'][i]>0:
        por.append(merge['denPor'][i])
    else:
        por.append(merge['neutronPor'][i])
```

```python
merge['Porosity']=por
```
